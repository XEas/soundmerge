```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
a = np.random.rand(10000)
b = np.random.rand(10000)
```

```python
plt.hist(a, bins=100);
```

```python
plt.hist(b, bins=100);
```

```python
plt.hist((a+b) / 2, bins=100);
```

```python
k1 = a * b
k2 = a * (1-b)
```

```python
plt.hist((k1+k2)/2, bins=100);
```

```python
plt.hist(a + b, bins=100);
```

```python
c = np.where(np.random.rand(10000) < 0.5, a, b)
```

```python
plt.hist(c, bins=100);
```

```python
dist1 = np.random.normal(0, 1, 10000)
```

```python
uniform_dist1 = stats.norm.cdf(dist1, loc=0, scale=1)
```

```python
plt.hist(uniform_dist1, bins=100);
```
