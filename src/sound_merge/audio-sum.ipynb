```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.random.rand(10000)
b = np.random.rand(10000)
```

```python
plt.hist(a, bins=100);
```

```python
plt.hist(b, bins=100);
```

```python
plt.hist((a+b) / 2, bins=100);
```

```python
k1 = a * b
k2 = a * (1-b)
```

```python
plt.hist((k1+k2)/2, bins=100);
```

```python
plt.hist(a + b, bins=100);
```

```python
c = np.where(np.random.rand(10000) < 0.5, a, b)
```

```python
plt.hist(c, bins=100);
```
